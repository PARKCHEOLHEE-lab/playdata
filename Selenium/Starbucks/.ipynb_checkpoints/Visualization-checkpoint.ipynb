```python
import pandas as pd
import folium
import json

sb_seoul = pd.read_excel("../Starbucks/df_starbucks_final.xlsx")
print(sb_seoul.head())
```

```python
# Starbucks location Visualization
# location=[37.5502, 126.982]

sb_map = folium.Map(location=[37.5502, 126.982], tiles='stamentoner', zoom_start=11)

for i in range(len(sb_seoul)):
    lat = sb_seoul.loc[i, "위도"]
    long = sb_seoul.loc[i, "경도"]
    
    folium.CircleMarker(
    location = [lat, long],
    fill = True,
    fill_color = "yellow",
    color = "yellow",
    fill_opacity = 0.5,
    radius = 5,
    width = 0.1).add_to(sb_map)

sb_map
```

```python
sb_type_map = folium.Map(location=[37.5502, 126.982], tiles='stamentoner', zoom_start=11)

for i in range(len(sb_seoul)):
    lat = sb_seoul.loc[i, "위도"]
    long = sb_seoul.loc[i, "경도"]
    s_type = sb_seoul.loc[i, "매장타입"]
    
    if s_type == "general":
        co = "red"
        size = 2
    else:
        co = "blue"
        size = 5
        
    folium.CircleMarker(
    location = [lat, long],
    fill = True,
    color = co,
    fill_color = co,
    radius = size    
    ).add_to(sb_type_map)

sb_type_map
```
